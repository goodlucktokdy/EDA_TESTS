````python
# 3-3: Boxplot
tmp_grouped = df.pivot_table(index = ['ê¶Œì—­','ìì¹˜êµ¬'], values = ['ì™¸êµ­ì¸ë¹„ìœ¨'])
````

````python
tmp_grouped.head()
````

````python
# tmp_groupedì—ì„œ 'ì™¸êµ­ì¸ë¹„ìœ¨' ë°ì´í„°ë¥¼ ì¶”ì¶œí•˜ê³ , ì¸ë±ìŠ¤ ë ˆë²¨ ê°’ì„ ê¸°ì¤€ìœ¼ë¡œ ê·¸ë£¹í™”
grouped_data = tmp_grouped['ì™¸êµ­ì¸ë¹„ìœ¨']
grouped_data = grouped_data.groupby(tmp_grouped.index.get_level_values(0))

# ë°•ìŠ¤í”Œë¡¯ì„ ê·¸ë¦¬ê¸° ìœ„í•œ ë°ì´í„° ë¦¬ìŠ¤íŠ¸ë¥¼ ìƒì„±
data_to_plot = [group for name, group in grouped_data]

# ë°•ìŠ¤í”Œë¡¯ ìƒì„±
fig, ax = plt.subplots(figsize = (12, 12))

# ë°•ìŠ¤í”Œë¡¯ ê·¸ë¦¬ê¸°
# ë§·í”Œë¡¯ë¦½ boxplotê·¸ë¦¬ë©´ ì†ì„±ë“¤ ë‹´ê¸´ ë”•ì…”ë„ˆë¦¬ë¥¼ ë°˜í™˜í•¨
bp = ax.boxplot(data_to_plot)

# Xì¶• ë ˆì´ë¸” ì„¤ì •
ax.set_xticklabels(grouped_data.groups.keys())

# ì¤‘ì•™ê°’ ë¼ì¸ ìƒ‰ ì„¤ì •
for median in bp['medians']:
    median.set_color('green')
    median.set_linewidth(1)
    
# quantile3 + 1.5 * IQRì§€ì ê³¼ ë°•ìŠ¤ ì—°ê²°í•´ ì£¼ëŠ” ìˆ˜ì—¼ê³¼
# quantile1 - 1.5 * IQRì§€ì ê³¼ ë°•ìŠ¤ ì—°ê²°í•´ ì£¼ëŠ” ìˆ˜ì—¼ í‘œì‹œí•´ì£¼ëŠ” ìƒ‰ ì„¤ì • 
for i in bp['whiskers']:
    i.set_color('#2596be')
    
# í…Œë‘ë¦¬ ìƒ‰ ì„¤ì •    
for box in bp['boxes']:
    box.set(color='#2596be')

# íƒ€ì´í‹€ ë° ë ˆì´ë¸” ì„¤ì •
ax.set_title('Boxplot grouped by ê¶Œì—­', pad = 40)
ax.set_xlabel('ê¶Œì—­')
plt.xticks(fontsize = 15, fontweight = 550)
plt.yticks(fontsize = 15, fontweight = 550)
plt.text(3, 8.5, 'ì™¸êµ­ì¸ë¹„ìœ¨', ha='center', va='bottom', fontsize = 12, fontweight = 550)
plt.grid(True)
# í”Œë¡¯ í‘œì‹œ
plt.show()
````
